```python
import gradio as gr
from googletrans import Translator
from gtts import gTTS
import tempfile
import os

translator = Translator()

LANGUAGES = {
    "English": "en",
    "Hindi": "hi",
    "Telugu": "te",
    "Tamil": "ta",
    "Kannada": "kn",
    "Malayalam": "ml",
    "Bengali": "bn",
    "Marathi": "mr",
    "Gujarati": "gu",
    "Punjabi": "pa",
    "Urdu": "ur",
    "French": "fr",
    "German": "de",
    "Spanish": "es",
    "Russian": "ru",
    "Chinese (Simplified)": "zh-cn",
    "Japanese": "ja",
    "Korean": "ko",
    "Arabic": "ar"
}

def translate_text(text, to_lang):
    if text.strip() == "":
        return "‚ùå Please enter text"

    lang_code = LANGUAGES.get(to_lang, "en")
    result = translator.translate(text, dest=lang_code)
    return result.text

def text_to_speech(text, to_lang):
    if text.strip() == "":
        return None

    lang_code = LANGUAGES.get(to_lang, "en")

    try:
        tts = gTTS(text=text, lang=lang_code)
        tmp = tempfile.NamedTemporaryFile(delete=False, suffix=".mp3")
        tts.save(tmp.name)
        return tmp.name
    except:
        return None

def translate_and_speak(text, to_lang):
    translated = translate_text(text, to_lang)
    audio = text_to_speech(translated, to_lang)
    return translated, audio

demo = gr.Interface(
    fn=translate_and_speak,
    inputs=[
        gr.Textbox(label="Enter Text", placeholder="Type here...", lines=4),
        gr.Dropdown(list(LANGUAGES.keys()), value="English", label="Translate To")
    ],
    outputs=[
        gr.Textbox(label="Translated Text"),
        gr.Audio(label="Read Aloud (Audio Output)")
    ],
    title="Language Translator (Notebook Version)",
    description="Works inside Jupyter/Colab/Kaggle. Translate + Read Aloud."
)

demo.launch()
```
